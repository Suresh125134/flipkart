```python
from tkinter import *
import ast
```

```python
root=Tk()

i=0
def display_numbers(num):
    global i
    display.insert(i,num)
    i+=1

def display_operators(op):
    global i
    display.insert(i,op)
    i+=len(op)

def clear_all():
    display.delete(0,END)

def undo():
    entry_str=display.get()
    clear_all()
    display.insert(i,entry_str[0:-1])

def calculate():
    entry_str=display.get()
    clear_all()
    try:
        node=ast.parse(entry_str,mode='eval')
        result=eval(compile(node,'<string>','eval'))
        display.insert(0,result)
    except Exception:
        display.insert(0,'Error')    

display=Entry(root)
display.grid(columnspan=6)

count=1
for i in range(3):
    for j in range(3):
        button=Button(root,text=count,width=4,height=2,command=lambda num=count:display_numbers(num))
        button.grid(row=i+1,column=j)
        count+=1

button=Button(root,text='0',width=4,height=2,command=lambda num=0:display_numbers(num))
button.grid(row=4,column=1)

operators=['+','-','*','/','**','(',')']
counter=0
for i in range(4):
    for j in range(2):
        if counter<len(operators):
            button=Button(root,text=operators[counter],width=4,height=2,
                          command=lambda op=operators[counter]:display_operators(op))
            if operators[counter]==')':
                button.grid(row=i+1,column=j+4)
            else:
                button.grid(row=i+1,column=j+3)
            counter+=1

Button(root,text='AC',width=4,height=2,command=clear_all).grid(row=4,column=0)

Button(root,text='Del',width=4,height=2,command=undo).grid(row=4,column=2)

Button(root,text='=',width=4,height=2,command=calculate).grid(row=4,column=3)

root.mainloop()
```

```python
root = Tk()

i=0
def get_number(num):
    global i
    display.insert(i,num)
    i+=1

def get_operations(operator):
    global i
    length=len(operator)
    display.insert(i,operator)
    i+=length

def clear_all():
    display.delete(0,END)

def calculate():
    expression=display.get()
    try:
        node=ast.parse(expression,mode='eval')
        result=eval(compile(node,'<string>','eval'))
        clear_all()
        display.insert(0,result)
    except Exception:
        clear_all()
        display.insert(0,'Error')

def undo():
    expression=display.get()
    if len(expression):
        last_char_deleted=expression[:-1]
        clear_all()
        display.insert(0,last_char_deleted)
    else:
        clear_all()
        display.insert(0,'')

display=Entry(root)
display.grid(row=1,columnspan=6)

counter=1
for x in range(3):
    for y in range(3):
        button=Button(root,text=counter,width=4,height=2,command=lambda text=counter:get_number(text))
        button.grid(row=x+2,column=y)
        counter+=1

button=Button(root,text='0',width=4,height=2,command=lambda :get_number(0))
button.grid(row=5,column=2)

count=0
operations=['+','-','*','/','**','(',')']
for x in range(4):
    for y in range(2):
        if count<len(operations):
            button=Button(root,text=operations[count],width=4,height=2,command=lambda text=operations[count]:get_operations(text))
            if operations[count]==')':
                button.grid(row=x+2,column=y+4)
            else:
                button.grid(row=x+2,column=y+3)
            count+=1

Button(root,text='AC',width=4,height=2,command=clear_all).grid(row=5,column=0)

Button(root,text='=',width=4,height=2,command=calculate).grid(row=5,column=3)

Button(root,text='Del',width=4,height=2,command=undo).grid(row=5,column=1)

root.title('Cal')

root.mainloop()
```
